```python
from stats import *
import matplotlib.pyplot as plt ; plt.rcParams["figure.figsize"] = (30,15)
```

```python
# Reading file from path and dropping na values
scorecard = pd.read_stata(os.getcwd() + '/data/scorecard.dta').dropna()

# Assigning the variable in question
feature = 'earnings_med'

# Assigning the null hypothesis to be conditional on having a degree
x1 = scorecard[scorecard['pred_degree_awarded_ipeds'] == 'Four-year+ degree'][feature]

# Assigning the alt hypothesis to be conditional on not having a degree
x2 = scorecard[scorecard['pred_degree_awarded_ipeds'] != 'Four-year+ degree'][feature]
```

```python
plt.plot(x1);
plt.plot(x2);
```

```python
CompareMeans(DescrStatsW(x1),  DescrStatsW(x2)).summary()
```
